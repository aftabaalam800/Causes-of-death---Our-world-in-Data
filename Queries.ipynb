```python
# Function to extract the cause of death with try-except
def extract_cause(col_name):
    try:
        # Try to split the column name and extract the cause of death
        return col_name.split('-')[1].strip()
    except IndexError:
        # If there's an IndexError, return the original column name
        return col_name

# Apply the function to rename columns
new_columns = [extract_cause(col) for col in df.columns]

# Rename the columns in the DataFrame
df.columns = new_columns

# Display the updated DataFrame
df.head()
```

### Q1. Retrieve the total number of deaths caused by "Meningitis" in Afghanistan over all available years.

```python
unique_cities_query1 = """SELECT SUM(Meningitis) FROM Death WHERE Entity = 'Afghanistan'; """
unique_cities_df1 = pd.read_sql_query(unique_cities_query1, engine)

print(unique_cities_df1)
```

### Q2.	List the distinct causes of death recorded in Afghanistan for the year 2007.

```python
unique_cities_query2 = """SELECT COUNT(DISTINCT Causes_name) FROM Death2 WHERE Entity = 'Afghanistan' ;"""
unique_cities_df2 = pd.read_sql_query(unique_cities_query2, engine)

print(unique_cities_df2)
```

### Q3. Find the year with the highest number of deaths due to "Cardiovascular diseases" in Afghanistan

```python
unique_cities_query3 = """SELECT Year,Cardiovascular_diseases FROM Death WHERE Entity = 'Afghanistan' ORDER BY Cardiovascular_diseases DESC LIMIT 1 ;"""
unique_cities_df3 = pd.read_sql_query(unique_cities_query3, engine)

print(unique_cities_df3)
```

### Q4. Identify the year and cause with the least number of deaths in Afghanistan.

### Q5. Count the total number of deaths from "Road injuries" from 1990 to 2000 in Afghanistan.

### Q7.	Determine the average number of deaths due to "Diarrheal diseases" over all years.

### Q8.	Group the data by cause and calculate the average deaths for each cause.

```python
unique_cities_query8 = """SELECT Causes_name,AVG(Death_numbers) FROM Death2 GROUP BY Causes_name ORDER BY AVG(Death_numbers) DESC; """
unique_cities_df8 = pd.read_sql_query(unique_cities_query8, engine)

print(unique_cities_df8)
```

### Q9.	Find the sum of deaths due to "Lower respiratory infections" for each year and list them in descending order.

```python
unique_cities_query9 = """SELECT Year,SUM(Lower_respiratory_infections) FROM Death GROUP BY Year ORDER BY SUM(Lower_respiratory_infections) DESC; """
unique_cities_df9 = pd.read_sql_query(unique_cities_query9, engine)

print(unique_cities_df9)
```
