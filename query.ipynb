```python
# Function to extract the cause of death with try-except
def extract_cause(col_name):
    try:
        # Try to split the column name and extract the cause of death
        return col_name.split('-')[1].strip()
    except IndexError:
        # If there's an IndexError, return the original column name
        return col_name

# Apply the function to rename columns
new_columns = [extract_cause(col) for col in df.columns]

# Rename the columns in the DataFrame
df.columns = new_columns

# Display the updated DataFrame
df.head()
```

### Q24.	Identify the causes that had a higher than average number of deaths in each year.

```python
unique_cities_query = """WITH YearlyAverages AS (
    SELECT 
        year, 
        AVG(death_Numbers) AS avg_deaths
    FROM 
        death
    GROUP BY 
        year
)
SELECT 
    d.year, 
    d.causes_name
FROM 
    death d
JOIN 
    YearlyAverages ya
ON 
    d.year = ya.year
WHERE 
    d.death_Numbers > ya.avg_deaths
GROUP BY 
    d.year, 
    d.causes_name
ORDER BY 
    d.year, 
    d.causes_name;



"""
unique_cities_df = pd.read_sql_query(unique_cities_query, engine)

print(unique_cities_df)
```

### Q25.	For each cause, find the year with the maximum number of deaths and the year with the minimum.

```python


unique_cities_query = """WITH MaxMinDeaths AS (
    SELECT 
        causes_name,
        year,
        SUM(Death_Numbers) AS total_deaths,
        MAX(SUM(Death_Numbers)) OVER (PARTITION BY causes_name) AS max_deaths,
        MIN(SUM(Death_Numbers)) OVER (PARTITION BY causes_name) AS min_deaths
    FROM 
        death
    GROUP BY 
        causes_name, year
)
SELECT 
    causes_name,
    year AS max_deaths_year,
    total_deaths AS max_deaths
FROM 
    MaxMinDeaths
WHERE 
    total_deaths = max_deaths

UNION ALL

SELECT 
    causes_name,
    year AS min_deaths_year,
    total_deaths AS min_deaths
FROM 
    MaxMinDeaths
WHERE 
    total_deaths = min_deaths
;



"""
unique_cities_df = pd.read_sql_query(unique_cities_query, engine)

print(unique_cities_df)
```

### Q28.	Identify causes where the deaths have consistently increased over three consecutive years.

```python
unique_cities_query = """
SELECT causes_name
FROM (
    SELECT causes_name, year, death_numbers,
           LAG(death_numbers, 1) OVER (PARTITION BY causes_name ORDER BY year) AS prev_year_1,
           LAG(death_numbers, 2) OVER (PARTITION BY causes_name ORDER BY year) AS prev_year_2
    FROM death
) AS subquery
WHERE prev_year_2 IS NOT NULL
  AND death_numbers > prev_year_1
  AND prev_year_1 > prev_year_2
GROUP BY causes_name;

"""
unique_cities_df = pd.read_sql_query(unique_cities_query, engine)

print(unique_cities_df)
```

### Q29.	Find the years where deaths due to "Alcohol use disorders" were more than those caused by "Drug use disorders".
